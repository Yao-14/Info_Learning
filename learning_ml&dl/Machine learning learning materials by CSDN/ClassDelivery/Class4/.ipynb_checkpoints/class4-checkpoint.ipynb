```python
#加载基础的库
import skimage
import numpy as np
import matplotlib.pyplot as plt
```

```python
#霍夫变换
```

```python
#根据第一节课学到的方法画一张测试图片

#加载画图需要模块
from skimage import draw
#画图

img=np.zeros([100,100])
#画一条线
h,w=draw.line(1,1,95,95)
img[h,w]=1



plt.imshow(img)


```

```python
#加载霍夫变换需要的模块
import skimage.transform as transform
```

```python
# hough线变换
h, theta, d = transform.hough_line(img)

print(h.shape)
print(theta.shape)
print(d.shape)
```

```python
#显示生成的结果图
plt.figure('hough_line',figsize=(8,8))
plt.subplot(121)
plt.imshow(img) 
plt.subplot(122)
plt.imshow(np.log(1+h))
plt.show()
```

```python
#显示生成的效果图

plt.figure('hough_line',figsize=(8,8))
plt.subplot(121)
plt.imshow(img) 
plt.subplot(122)
row1, col1 = img.shape
for _, angle, dist in zip(*transform.hough_line_peaks(h, theta, d)):
    y0 = (dist - 0 * np.cos(angle)) / np.sin(angle)
    y1 = (dist - col1 * np.cos(angle)) / np.sin(angle)
    plt.plot((0, col1), (y0, y1), '-r')
plt.axis((0,col1,row1,0))
plt.show()
```

```python
#霍夫圆检测
```

```python
#画图
img=np.zeros([200,200])
#画一条线
h,w=draw.line(1,1,10,150)
img[h,w]=1

#画一圆
h, w = draw.circle_perimeter(150, 50, 30) 
img[h,w]=1

#h,w=draw.circle(150,150,15) #大圆
#img[h,w]=1

#画一个正方形
img[40:60,60:80]=1  #矩形

plt.imshow(img)
```

```python
#开始霍夫圆检测

#首先设定检测的半径范围
radii=np.arange(5,50,5)
print(radii)
#霍夫检测
circle=transform.hough_circle(img,radii)
print(circle.shape)
```

```python
#展示识别效果,看不同半径范围检测到的最大值

r=2
print(np.max(circle[r,:,:]))
plt.imshow(circle[r,:,:])
```

```python
#把检测到的圆画出来

#设置检测到的半径
r=5
#找到最大值的位置
import skimage.feature as feature
peak =feature.peak_local_max(circle[r,:,:], num_peaks=1)

print(peak)

#画圆
img2=img.copy()
cx, cy =draw.circle_perimeter(peak[0][0], peak[0][1], radii[r])
img2[cx,cy]=255

plt.figure("detect_circle",figsize=(8,8))
plt.subplot(121)
plt.imshow(img) 
plt.subplot(122)
plt.imshow(img2)
plt.show()

```
